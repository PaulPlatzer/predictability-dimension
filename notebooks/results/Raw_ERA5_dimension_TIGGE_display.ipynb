# Raw ERA5 Dimension and TIGGE Spread

This notebook displays the current raw-field workflow. The local dimensions are computed from the raw 2-degree ERA5 z500 field, then aligned with TIGGE ensemble spread for visual comparison.

Heavy computations should be run from the command line; this notebook is intentionally limited to loading NetCDF outputs, plotting figures, and recording scientific interpretation.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import xarray as xr

from predictability_dimension.plotting import (
    align_spread_and_dimension,
    plot_dimension_vs_spread,
    plot_dimension_vs_spread_growth,
)
```

## Inputs

`data/dim_raw_era5_2deg.nc` is the expected production output. If it is not present yet, the notebook falls back to the smoke-test output produced while validating the code.

```python
spread_path = Path("data/TIGGEspread.nc")
dimension_path = Path("data/dim_raw_era5_2deg.nc")
smoke_dimension_path = Path("outputs/smoke/dim_raw_era5_2deg_all_estimators_smoke.nc")
figure_dir = Path("figures/generated")
figure_dir.mkdir(parents=True, exist_ok=True)

if not dimension_path.exists():
    dimension_path = smoke_dimension_path

print(f"Spread file: {spread_path}")
print(f"Dimension file: {dimension_path}")
```

```python
spread = xr.open_dataarray(spread_path)
dimension = xr.open_dataset(dimension_path)

spread, dimension = align_spread_and_dimension(spread, dimension)
display_steps = spread.step.isel(step=range(2, 8))
display_spread = spread.sel(step=display_steps)

dimension
```

## Dimension vs TIGGE Spread

```python
fig = plot_dimension_vs_spread(
    display_spread,
    dimension,
    output=figure_dir / "raw_era5_spread_dimension.png",
    title="Raw ERA5 dimension vs TIGGE spread",
)
plt.show()
```

## Dimension vs Spread Growth

```python
fig = plot_dimension_vs_spread_growth(
    spread,
    dimension,
    reference="previous",
    steps=display_steps,
    output=figure_dir / "raw_era5_spread_growth_previous_dimension.png",
)
plt.show()
```

```python
fig = plot_dimension_vs_spread_growth(
    spread,
    dimension,
    reference="initial",
    steps=display_steps,
    output=figure_dir / "raw_era5_spread_growth_initial_dimension.png",
)
plt.show()
```
